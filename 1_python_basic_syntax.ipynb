## Variables and Data Types

**Variables**

A variable in Python is a name that acts as a reference or a container to store data values. Variables allow you to work with data by giving it a meaningful name that represents the value or data structure.

Rules for Naming Variables:
* Must start with a letter (a-z, A-Z) or an underscore (_).
* Cannot start with a digit.
* Can contain letters, digits, and underscores (e.g., my_variable).
* Cannot use Python keywords or reserved words (e.g., if, else, while, etc.).
* Should not contain special characters (e.g., @, #, $, etc.).

```python
name = "Alice" # String
age = 25 # Integer
height = 5.5 # Float
is_active = True # Boolean
```

**Primitive Data Types**

Primitive data types (also called basic or fundamental data types) are the building blocks of data manipulation in Python. They represent simple, atomic values.

Common Primitive Data Types in Python:
1. Integer (**int**) - Represents whole numbers (positive, negative, or zero).
2. Floating-point (**float**) - Represents real numbers with decimal points.
3. String (**str**) - Represents a sequence of characters enclosed in single (') or double (") quotes.
4. Boolean (**bool**) - Represents one of two values: True or False.

*Characteristics of Primitive Data Types:*

* **Immutable**: Primitive data types cannot be modified after creation. For
example, changing a string or an integer creates a new object in memory.
* **Simple Values**: These types store a single value and are not collections or complex structures.

```python
a = 10 # Integer
b = 3.14 # Float
c = "Machine Learning" # String
d = True # Boolean
```

**Collection Data Types**

Collection data types in Python are used to group multiple items into a single object. These types allow you to store, manipulate, and manage collections of data. They are highly flexible and come in various forms based on use cases.

**Types of Collection Data Types**

1.**List (list)**

**Definition:** An ordered, mutable collection of items that can hold different types of data.

**Key Features:**
* Ordered (items maintain their position).
* Allows duplicates.
* Mutable (can be modified after creation).

2.**Tuple (tuple)**

**Definition:** An ordered, immutable collection of items.

**Key Features:**
* Ordered (items maintain their position).
* Allows duplicates.
* Immutable (cannot be changed after creation).

3.**Set (set)**

**Definition:** An unordered, mutable collection of unique items.

**Key Features:**
* Unordered (items do not maintain their position).
* No duplicates allowed.
* Mutable (but the elements themselves must be immutable).

4.**Dictionary (dict)**

**Definition:** An unordered, mutable collection of key-value pairs.

**Key Features:**
* Keys must be unique and immutable.
* Values can be any data type and mutable.
* Efficient for lookups by key.

5.**String (str)**

**Definition:** Although technically a primitive type, strings in Python can be treated as collections of characters.

**Key Features:**
* Ordered (characters maintain their position).
* Immutable (cannot change the string after creation).


**Use Cases of Collection Types**
* **Lists**: When you need an ordered collection that can grow or shrink dynamically (e.g., shopping cart items).
* **Tuples**: When you need a fixed, ordered collection (e.g., RGB color codes).
* **Sets**: When you need unique items without caring about order (e.g., list of unique email addresses).
* **Dictionaries**: When you need to associate keys with values for fast lookups (e.g., configuration settings).

```python
my_list = [1, 2, 3, 4] # List
my_tuple = (1, 2, 3, 4) # Tuple
my_set = {1, 2, 3, 4}  # Set
my_dict = {"key1": "value1", "key2": "value2"}
```

**Feature Engineering**

Feature Engineering is the process of preparing data to improve machine learning model performance by creating, transforming, and selecting features.

**Key Steps:**
* Feature Selection: Identify and keep only relevant features (e.g., using correlation, feature importance).
* Feature Transformation: Modify features to make them more useful (e.g., scaling, log transform).
* Feature Creation: Generate new features (e.g., combining features, extracting time-based information).
* Handle Missing Data: Fill missing values using mean, median, mode, or remove rows/columns.
* Encode Categorical Variables: Convert categories to numerical values (e.g., one-hot encoding, label encoding).
* Handle Outliers: Treat extreme values (e.g., capping, removing).
* Dimensionality Reduction: Reduce the number of features (e.g., PCA).


**Tools:**
* Libraries: Pandas, scikit-learn, NumPy, FeatureTools.
* Techniques: Scaling, encoding, imputation, interaction terms.

Example:

Normalizing data using MinMaxScale from scikit-learnr

```python
from sklearn.preprocessing import MinMaxScaler
data = [[-1, 2], [0, 0], [1, -1]]
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)
print(scaled_data)
```
